## Arranque del servidor

En una terminal antes de la demo:

```bash
source .venv/bin/activate
source scripts/load_model_from_akv.sh
python -m microharness.server
```

````python
import httpx

base_url = 'http://127.0.0.1:8000'
try:
    print(httpx.get(f'{base_url}/health', timeout=5).json())
except Exception as exc:
    print(f'Plan B: servidor no disponible ({exc})')
````

````python
prompt = 'Explícame Agent Harness en 5 bullets para una audiencia técnica.'
try:
    response = httpx.post(f'{base_url}/api/chat', json={'prompt': prompt}, timeout=90)
    print(response.json()['response'])
except Exception as exc:
    print(f'Plan B: usar working/fallback/expected_response.md ({exc})')
````

## AG-UI

El endpoint principal para streaming es `/agent`. El endpoint `/api/chat` queda como fallback REST estable para una demo cronometrada.
