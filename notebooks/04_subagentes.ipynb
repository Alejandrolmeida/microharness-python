# 04 · Workflows y agentes componibles en MAF

Para mostrar subagentes de forma MAF-native usamos `@workflow` y `FunctionalWorkflowAgent`. Este ejemplo evita depender del modelo y enseña la composición con primitivas del framework.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT / 'src'))
```

```python
import agent_framework as af


@af.workflow(name="microharness_review")
async def review_workflow(task: str) -> str:
    checks = [
        f"Arquitectura: separar responsabilidades para {task}.",
        "Fiabilidad: mantener trazas compactas y respuestas reproducibles.",
        "Seguridad: pedir aprobación humana para acciones sensibles.",
    ]
    return "\n".join(f"- {check}" for check in checks)


workflow_agent = af.FunctionalWorkflowAgent(review_workflow, name="MicroHarnessWorkflowAgent")
{
    "workflow_type": type(review_workflow).__name__,
    "agent_type": type(workflow_agent).__name__,
    "agent_name": workflow_agent.name,
}
```

```python
result = await workflow_agent.run("revisar la variante MAF-native")
print(result.text)
```
