## 3:45–4:40 · Ejecución

Ejecutar una llamada fija y mostrar respuesta, artefacto y sesión.

```python
from microharness.memory import (  # noqa: E402
    read_fallback_response,
    read_session_state,
    write_demo_summary,
)

prompt = (
    'Explícame Agent Harness en 5 bullets para una audiencia técnica. '
    'Usa el contexto disponible y separa Agent Loop, Tools, Memory, Planning y Permissions.'
)
try:
    response = httpx.post(
        'http://127.0.0.1:8000/api/chat',
        json={'prompt': prompt},
        timeout=90,
    )
    payload = response.json()
    print(payload['response'])
    print('Artefacto:', payload['artifact_path'])
    print('Sesión:', payload['session_state'])
except Exception:
    summary = read_fallback_response()
    print(summary)
    print(write_demo_summary(prompt, summary))
    print(read_session_state())
```
