Primer paso: comprobar la configuración del modelo y localizar las primitivas nativas de Microsoft Agent Framework que usará la demo.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT / 'src'))
ROOT
```

```python
import agent_framework as maf

from microharness.config import load_settings
from microharness.maf_native import MAF_NATIVE_DIR

settings = load_settings()
maf_primitives = [
    "Agent",
    "AgentSession",
    "ContextProvider",
    "FileHistoryProvider",
    "MemoryContextProvider",
    "TodoProvider",
    "ToolApprovalMiddleware",
    "AgentMiddleware",
    "FunctionMiddleware",
]

{
    "model_configured": settings.is_model_configured,
    "model": settings.model,
    "route": "base_url" if settings.uses_openai_compatible_url else "azure_endpoint",
    "maf_native_dir": str(MAF_NATIVE_DIR.relative_to(ROOT)),
    "maf_primitives_available": {name: hasattr(maf, name) for name in maf_primitives},
}
```
