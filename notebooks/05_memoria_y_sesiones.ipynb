# 05 · Memoria y sesiones nativas de MAF

La variante MAF-native usa `AgentSession` para la continuidad, `FileHistoryProvider` para historial, `MemoryContextProvider` para memoria experimental y `TodoProvider` para planificación persistida.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT / 'src'))
```

```python
from agent_framework import AgentSession

from microharness.maf_native import build_native_context_providers

session = AgentSession(session_id="notebook-maf-native")
session.state["microharness"] = "notebook"
providers = build_native_context_providers()

{
    "session_id": session.session_id,
    "state": session.state,
    "providers": [(type(provider).__name__, provider.source_id) for provider in providers],
}
```

```python
from microharness.maf_native import MAF_NATIVE_DIR

files = [
    path.relative_to(ROOT).as_posix()
    for path in MAF_NATIVE_DIR.rglob("*")
    if path.is_file()
]
sorted(files)[:20]
```

La clave práctica: `MemoryFileStore` y `TodoFileStore` necesitan un identificador de owner en `session.state`. En la app lo inicializa `HarnessKnowledgeProvider` antes de que los providers de memoria/todos lean disco.
